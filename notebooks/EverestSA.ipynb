```python
# Age groups
everest['age_group'] = pd.cut(everest['Age'], 
                                    bins=[0, 30, 40, 50, 60, 100],
                                    labels=['Under 30', '30-40', '40-50', '50-60', '60+'])
print("\nDeaths by Age Group:")
print(everest['age_group'].value_counts().sort_index())
```

```python
# Remove the 'm' and convert to numeric
everest['Location_in_Meters'] = (
    everest['Location_in_Meters']
    .str.replace('m', '', regex=False)
    .astype(float)
)
# Deaths by altitude zone
def categorize_altitude(alt):
    if pd.isna(alt):
        return 'Unknown'
    elif alt < 7000:
        return 'Below 7000m'
    elif alt < 8000:
        return '7000-8000m'
    elif alt < 8500:
        return '8000-8500m (Death Zone)'
    else:
        return 'Above 8500m (Summit Zone)'

everest['altitude_zone'] = everest['Location_in_Meters'].apply(categorize_altitude)
print("\nDeaths by Altitude Zone:")
print(everest['altitude_zone'].value_counts())

# Cause of death by altitude zone
print("\nCause of Death by Altitude Zone:")
print(pd.crosstab(everest['altitude_zone'], everest['Cause of death']))
```

```python
# Gender vs Cause of Death
print("\nGender vs Cause of Death:")
print(pd.crosstab(everest['Gender'], everest['Cause of death']))
```

```python
# Experience Level vs Cause of Death
print("\nExperience Level vs Cause of Death:")
print(pd.crosstab(everest['Experience_Level'], everest['Cause of death']))
```

```python
# Independent vs Expedition Company
print("\nExpedition Type:")
everest['is_independent'] = everest['Expedition_Company'].str.contains('Independent', na=False)
print(everest['is_independent'].value_counts())
```

```python
# Chi-square test: Is experience level independent of cause of death?
contingency = pd.crosstab(everest['Experience_Level'], everest['Cause of death'])
chi2, p_value, dof, expected = stats.chi2_contingency(contingency)
print(f"\nExperience Level vs Cause of Death (Chi-square test):")
print(f"Chi-square: {chi2:.3f}, P-value: {p_value:.3f}")
if p_value < 0.05:
    print("Experience level and cause of death are related")
else:
    print("No significant relationship")
```

```python
# 1. Basic breakdown
print("\n1. EXPEDITION COMPANY BREAKDOWN")
print("-" * 80)
everest['Expedition_Company'] = everest['Expedition_Company'].apply(
    lambda x: 'Independent/Private' if pd.isna(x) or 'Independent' in str(x) or 'Private' in str(x) 
    else 'Commercial Company'
)
print(everest['Expedition_Company'].value_counts())
print(f"\nPercentage independent: {(everest['Expedition_Company'] == 'Independent/Private').mean()*100:.1f}%")
```

```python
# 2. Deaths by specific companies
print("\n2. DEATHS BY EXPEDITION COMPANY (Top 10)")
print("-" * 80)
company_deaths = everest['Expedition_Company'].value_counts().head(10)
print(company_deaths)
```

```python
# 3. Independent vs Commercial: Key Metrics
print("\n3. INDEPENDENT VS COMMERCIAL COMPARISON")
print("-" * 80)
comparison = everest.groupby('Expedition_Company').agg({
    'Age': ['mean', 'median', 'std'],
    'Location_in_Meters': ['mean', 'median'],
    'Gender': lambda x: (x == 'Male').sum() / len(x) * 100  # % male
}).round(2)
print(comparison)
```

```python
# 4. Cause of Death by Expedition Type
print("\n4. CAUSE OF DEATH: INDEPENDENT VS COMMERCIAL")
print("-" * 80)
cause_by_type = pd.crosstab(everest['Expedition_Company'], 
                             everest['Cause of death'], 
                             normalize='index') * 100
print(cause_by_type.round(1))
```

```python
# 5. Experience Level by Expedition Type
print("\n5. EXPERIENCE LEVEL: INDEPENDENT VS COMMERCIAL")
print("-" * 80)
exp_by_type = pd.crosstab(everest['Expedition_Company'], 
                          everest['Experience_Level'],
                          normalize='index') * 100
print(exp_by_type.round(1))
```

```python
# 6. Statistical Test: Are independent climbers dying at different altitudes?
from scipy import stats

independent_alt = everest[everest['Expedition_Company'] == 'Independent/Private']['Location_in_Meters'].dropna()
commercial_alt = everest[everest['Expedition_Company'] == 'Commercial Company']['Location_in_Meters'].dropna()

if len(independent_alt) > 0 and len(commercial_alt) > 0:
    t_stat, p_value = stats.ttest_ind(independent_alt, commercial_alt)
    print(f"\n6. STATISTICAL TEST: ALTITUDE DIFFERENCE")
    print("-" * 80)
    print(f"Independent avg altitude: {independent_alt.mean():.0f}m")
    print(f"Commercial avg altitude: {commercial_alt.mean():.0f}m")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.3f}")
    if p_value < 0.05:
        print("✓ Statistically significant difference in death altitude")
    else:
        print("✗ No significant difference in death altitude")
```

```python
# 7. Top Commercial Companies - Detailed Analysis
print("\n7. TOP COMMERCIAL COMPANIES DETAILED ANALYSIS")
print("-" * 80)
commercial_only = everest[everest['Expedition_Company'] == 'Commercial Company']
top_companies = commercial_only['Expedition_Company'].value_counts().head(5).index

for company in top_companies:
    company_data = commercial_only[commercial_only['Expedition_Company'] == company]
    print(f"\n{company}:")
    print(f"  Total deaths: {len(company_data)}")
    print(f"  Avg age: {company_data['Age'].mean():.1f}")
    print(f"  Most common cause: {company_data['Cause of death'].mode()[0] if len(company_data['Cause of death'].mode()) > 0 else 'N/A'}")
    print(f"  Avg altitude: {company_data['Location_in_Meters'].mean():.0f}m")
```
